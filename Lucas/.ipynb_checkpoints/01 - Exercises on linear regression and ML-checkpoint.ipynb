* Study the effect of the number of examples N (from 2 to 200) and of noise dispersion on the coefficient estimates. Are the estimates of the coefficients stable?
* Study the effect of N on R2 (R square).
* Increase the dispersion of the noise. What is the effect on R2?
* Produce plots with the variation of R2 and of the parameters with the number of examples (up to 5000) and with the dispersion of the noise (sd between 0 and 5).


```python
n_s = [2, 20, 100, 200]
noise_dispersion_s = [1, 2, 3]

fig, axs = plt.subplots(len(n_s), len(noise_dispersion_s), figsize=(12, 12))
# fig.suptitle('Vertically stacked subplots')

for i in range(len(n_s)):
    for j in range(len(noise_dispersion_s)):
        
        n = n_s[i]
        noise_dispersion = noise_dispersion_s[j]
        np.random.seed(23)
        
        x=np.arange(n)
        X=x[:,np.newaxis] # newaxis gives X the shape of a matrix. Try without
        
        y=x*2 + 3 + np.random.randn(n) * 10**noise_dispersion

        model = LinearRegression(fit_intercept=True)
        model.fit(X,y)

        xfit = np.array([0,max(x)])
        yfit = model.predict(xfit[:,np.newaxis])

        axs[i, j].plot(x, y, 'o', color='blue', markersize=1);
        if i==len(n_s)-1:
            axs[i, j].set_xlabel('Noise Dispersion ' + str(10**noise_dispersion))
            
        if j == 0:
            axs[i, j].set_ylabel('number of examples '+ str(n))

        axs[i, j].plot(xfit, yfit,color='red');
```
